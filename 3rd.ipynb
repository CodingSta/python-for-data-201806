```python
def mycumsum(column_series):
    result = 0
    mylist = []
    for value in column_series:
        result += value
        mylist.append(result)
    return pd.Series(mylist, index=column_series.index)

df.apply(mycumsum)
```

```python
def myfn(column_series):
    return column_series.max() - column_series.min()

df.apply(myfn)
```
